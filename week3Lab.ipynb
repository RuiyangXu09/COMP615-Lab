```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

NZ_Rain = pd.read_csv("Rainfall_Data.csv")
NZ_Rain.head()
NZ_Rain.describe()
```

```python
from datascience import *
"""
在python中 make_array()中的index数值并不代表csv文件中的行数rows，以输入0为例，实际输出的行，是excel列表中的第2行
也就是说，实际输出的列表行数为make_array(indexNumber + 2),
举例如果我们想要输出第50行的内容,则应该编写代码make_array(48),
同理 make_array(100)实际输出的行数为excel列表的第102行
"""
NZ_Rain.take(make_array(3,18,100))
```

```python
"""
sort_values()函数原理类似于SQL中的order by，
可以将数据集依照某个字段中的数据进行排序，该函数即可根据指定列数据也可根据指定行的数据排序
DataFrame.sort_values(by=‘##’,axis=0,ascending=True, inplace=False, na_position=‘last’)
参数说明:
by ->指定列名(axis=0或’index’)或索引值(axis=1或’columns’)
axis -> 若axis = 0或’index’，则按照指定列中数据大小排序；若axis = 1或’columns',
则按照指定索引中数据大小排序，默认axis=0
ascending -> 是否按指定列的数组升序排列，默认为True，即升序排列
inplace -> 是否用排序后的数据集替换原来的数据，默认为False，即不替换
na_position -> {‘first’,‘last’}，设定缺失值的显示位置
"""
Sorted_siteName = NZ_Rain.sort_values(by = ['site'], ascending = False) #以site中的字母顺序进行排列的同时 默认为True，即升序排列, 将ascending函数设为False，即降序排列
Sorted_siteName.take(make_array(3,18,100))
```

```python
"""
sort_values(by = ['xxxxxx','xxxxxx',....... 'xxxxx'])可以定义多个数据的排列顺序
"""

Sorted_Date = NZ_Rain.sort_values(by=['period_start', 'period_end'], ascending = True) ##
Sorted_Date.take(make_array(3,18,100))
```

```python
#TO DO 1
#另一种日期排列顺序方法
import pandas as pd
import datetime
#导入数据集
data = pd.read_csv("Rainfall_Data.csv")

#标准化日期，获取时间的日月年
def change_date(date):
    date = datetime.datetime.strptime(date, "%d/%m/%Y") #把日期标准化，转化结果为 2017/1/30 转化为2017-01-30 00:00:00
    date = str(date)#上一步把date转化为时间格式，要把date转回str格式
    return date[:10] #只获取年月日，即位置10之前的字符串

data['period_start'] = data['period_start'].map(change_date) #用上面定义过的change_date函数 处理perio_start这一个事件列
data = data.sort_values(by = ['period_start', 'agent_number']) #按 period_start 和 agent_number进行升序排列
#筛选输出
data.filter(["agent_number", "site", "season", "precipitation", "period_start"])
```

```python
"""
filter()函数用于过滤序列，过滤掉不符合条件的元素，返回符合条件的元素组成新列表
filter(function, iterable) 其中function为函数，iterable为序列
"""
NZ_Rain.filter(["agent_number", "site", "season", "precipitation", "period_start"])
```

```python
#创建一个列表，该列表中仅包含具有关键词的数据，
#在这行代码中'Wangarei'就是在列表中不存在的数据，因此只有'Tara Hills'的数据被存入
sitename = ['Tara Hills', 'Wangarei']
"""
isin()是series中的一个函数，用于查看某列中是否包含某个字符串，返回值为布尔Series，来表明每一行的情况
在该行代码中一个新的数组 mysites[]创建用于存储新的数据，
使用isin()函数判断'site'列的数值是否存在于sitename数组中，如果该列表中由'site'列的数据，则返回true
"""
mysites = NZ_Rain[NZ_Rain['site'].isin(sitename)]
##mysites[]数组，使用filter()函数来过滤包含以下几列数据，并显示
mysites.filter(["agent_number", "site", "season", "precipitation", "period_start"])
```

```python
"""
between() 方法用于 Series 检查哪个值处在第一个、和第二个参数之间,
Series.between(left, right, inclusive=True)
将会返回一个布尔序列，对位于2个参数之间的值 left <= series <= right，返回 True，否则返回False
"""
#寻找 precipitation 数值位于100 到 200 的数据，并存入数组
mysites = NZ_Rain[NZ_Rain['precipitation'].between(100,200)]
#使用sort_values()函数来对该数组进行排序，以precipitation的数据进行倒序输出
mysites = mysites.sort_values(by = 'precipitation', ascending = False)

mysites.filter(["agent_number", "site", "season", "precipitation", "period_start"])
```

```python
#TO DO 2
#生成一个cityName数组，仅存入Aucklan 和 Hamilton两列的数据
cityName = ['Auckland', 'Hamilton']
#判断这两列的数据Auckland 和 Hamilton 是否在NZ_Rain 大数组中的 site列中存在
newList = NZ_Rain[NZ_Rain['site'].isin(cityName)]
#将以precipitation进行升序排列
newList = newList.sort_values(by= 'precipitation')
#排序后生成了一个新的newList数组，因此，不在传递NZ_Rain数组，改为传递新的newList数组
newList = newList[newList['precipitation'].between(500,999)]
#仅显示包含以下字段的数据
newList.filter(["agent_number", "site", "season", "precipitation", "period_start"])
```

```python
#sample()函数, 从列表中随机抽取n个元素，并将n个元素生以list形式返回
NZ_Rain.sample()
#随机抽取列表中的10个元素
NZ_Rain.sample(10)
```

```python
"""
sample()函数中的参数frac是要返回的比例，比如一个dataframe中有10行数据，只返回其中的30%,那么frac=0.3
"""
NZ_Rain.sample(frac=0.7) #返回dataframe中70%的数据
```
