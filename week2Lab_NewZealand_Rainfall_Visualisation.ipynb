```python
##导入用于数据操作的库
import numpy as np ##导入线性代数
import pandas as pd  ##处理数据，进行csv类型文件的I/O
##导入用于数据可视化的库
import matplotlib.pyplot as plt
"""
Seaborn其实是在matplotlib的基础上进行了更高级的API封装，
从而使得作图更加容易，在大多数情况下使用seaborn就能做出很具有吸引力的图
seaborn库中存有大量的图表类型，kdeplot( )密度曲线图，箱型图 boxplot( )，联合分布jointplot( )，热点图heatmap( ) 
线型图lineplot（），条状图barplot（）等等
"""
import seaborn as sns
%matplotlib inline
##读取CSV类型文件到DataFrame
NZ_Rain = pd.read_csv("Rainfall_Data.csv")
NZ_Rain.head()
NZ_Rain.describe()
##当需要同时显示一个或多个数值变量时，线图非常有用。以下为显示部分时间的降雨量线型图 代码 
"""matplotlib 中设置图形大小的语句如下：

 fig = plt.figure(figsize=(a, b), dpi=dpi)

其中：

    figsize 设置图形的大小，a 为图形的宽， b 为图形的高，单位为英寸
    dpi 为设置图形每英寸的点数
也就是说，同一像素尺寸的图形（比如 1200 * 600）可以有任意种不同的组合，比如：

figsize=(15,7.5), dpi= 80
figsize=(12,6)  , dpi=100
figsize=( 8,4)  , dpi=150
figsize=( 6,3)  , dpi=200
"""
plt.figure(figsize = (20,10))
plt.title("Total precipitation per years")
sns.lineplot(data = NZ_Rain, x = "period_start", y = "precipitation")
```

```python
"""
 为了描述离散类别之间的比较，可以使用条形图。
 在条形图中，将比较的特定类别与对应于这些类别的测量值进行对比。以下代码绘制了特定于现场的降雨量测量值： 
"""
plt.figure(figsize = (20,10))
plt.title("Total precipitation per Location")
sns.barplot(data = NZ_Rain, x = "precipitation", y = "site")  ##注意，所有折线图，条状图等一定是通过dataFrame传过来的
```

```python
plt.figure(figsize = (20,10))
plt.title("New Zealand anomaly per participation")
sns.scatterplot(data = NZ_Rain, x = "precipitation", y = "anomaly")
```
