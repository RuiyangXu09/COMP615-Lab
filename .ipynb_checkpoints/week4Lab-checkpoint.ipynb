```python
my_mpi_ds = pd.read_csv("MPI_Dataset.csv")
my_mpi_ds.head()
```

```python
my_mpi_ds = my_mpi_ds.drop(['ISO','Headcount Ratio Urban','Intensity of Deprivation Urban',
                            'Headcount Ratio Rural','Intensity of Deprivation Rural'], axis=1)

my_mpi_ds.rename(columns={'Country':'country', 'MPI Urban':'mpi_urban', 'MPI Rural':'mpi_rural'}, inplace=True)

my_mpi_ds.head(3)
```

```python
combined = pd.merge(
    Pov_data,
    my_mpi_ds,
    on='country',
    how='inner'
)

combined.head()
```

```python
#TO DO 2
"""
According to the scatter martix. the child_mort has a very strong positive multicollinearity with these elements:
mpi_urban, mpi_rural,totalfer,and strong negative multicollinearity with  and life_expec, gdp and income
for the health, inflation and imports, these three elements have a very weak multicollinearity with different elements.
for the income. the mpi_urban, mpi_rural and total_fer and child_mort have a strong negative multicollinearity and
strong positive multicollinearity with gdp and life_expec. for the mpi_urban. it has a very strong multicollinearity
with the mpi_rural 0.94 and 0.85 with total_fer and 0.82 with child_mort.
"""
combined.corr()

pd.plotting.scatter_matrix(combined, figsize=[20,20])
plt.show()

plt.figure(figsize=(15,10))
sns.heatmap(combined.corr(method='pearson', min_periods=1), annot=True)
```

```python
reg = linear_model.LinearRegression()

import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
#使用child_mort 作为X，预测mpi_urban(国内城市地区的多维贫困指数)作为Y
"""
左上图：表示预测值跟真实值之间的差值
右上图：表示x1的残差图。X坐标为x1的值，Y坐标表示x1对应的残差值，也就是左上图中蓝点跟黄点之间的距离差。
点离中间的直线越近表示模型更加有效
左下图：显示的是考虑新增其他自变量时，X与y之间的关系。
右下图：显示的是左下图图像的延申，反映了考虑新增其他自变量后反应两者关系的直线将如何变化
"""
model = ols('mpi_urban ~ child_mort', data=combined).fit()

print(model.summary())

fig = plt.figure(figsize=(12,8))

fig = sm.graphics.plot_regress_exog(model, 'child_mort', fig=fig)
#儿童死亡率越高，代表国内城市地区的贫困指数越高，儿童死亡率与国内城市地区的多维贫困指数有强多重共线性关系
```

```python
#TO DO 3:

#以child_mort 预测 mpi_rural(全国农村地区的多维贫困指数)
model2 = ols('mpi_rural ~ child_mort', data=combined).fit()

print(model2.summary())

fig2 = plt.figure(figsize=(12,8))

fig2 = sm.graphics.plot_regress_exog(model2, 'child_mort', fig=fig2)
#儿童死亡率越高，代表全国农村地区的多维贫困指数越高，儿童死亡率与全国农村地区的多维贫困指数有强多重共线性关系
```

```python
#TO DO 3:

#以income 预测 gdpp
model3 = ols('gdpp ~ income', data=combined).fit()

print(model3.summary())

fig3 = plt.figure(figsize=(12,10))

fig3 = sm.graphics.plot_regress_exog(model3, 'income', fig=fig3)
#国民的个人收入越高，gdp的指数也就越高，收入与gdp有着强多重共线性关系
```

```python
#TO DO 3:

#使用儿童死亡率，预测预期寿命life_expec
model4 = ols('life_expec ~ child_mort', data=combined).fit()

print(model4.summary())

fig4 = plt.figure(figsize=(12,8))

fig4 = sm.graphics.plot_regress_exog(model4, 'child_mort', fig=fig4)
#儿童死亡率越低，则预期寿命越高，也就是说预期寿命和儿童死亡率有着负相关，但是也有着强多重共线性关系
```

```python
#TO DO 3:

#使用全国农村地区的多维贫困指数（mpi_rural）预测收入（income）
model5 = ols('income ~ mpi_rural', data=combined).fit()

print(model5.summary())

fig5 = plt.figure(figsize=(12,8))

fig5 = sm.graphics.plot_regress_exog(model5, 'mpi_rural', fig=fig5)
#全国农村地区的多维贫困指数越低，则个人收入越高，mpi_rural与income有着负相关，但是也有强多重共线性关系
```
